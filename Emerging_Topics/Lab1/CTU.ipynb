```python
import numpy as np

#import the two datasets from sample_data
import pandas as pd

df1 = pd.read_csv('/content/CTU13_Attack_Traffic.csv')  # Replace with the actual file name
df2 = pd.read_csv('/content/CTU13_Normal_Traffic.csv')

#Merge the two dataframes
data = pd.concat([df1, df2], ignore_index=True)

#Display the merged dataframe
print(data.head())
```

```python
#check for missing values
data.isnull().sum()
```

## The data has no missing values so it does not need preprocessing

```python
#Print statistics
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr_matrix = data.corr()

# Set up figure size
plt.figure(figsize=(14, 10))

# Create heatmap with improvements
sns.heatmap(corr_matrix,
            annot=True,  # Set to False if too cluttered
            cmap='coolwarm',
            linewidths=0.5,
            fmt=".2f",  # Limit decimal places
            annot_kws={"size": 6})  # Reduce annotation font size

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title('Correlation Matrix')
plt.show()
```

```python
reduced_features = [
    'Flow Duration', 'Tot Fwd Pkts', 'TotLen Fwd Pkts',
    'Flow Bytes/s', 'Flow Pkts/s',
    'Fwd IAT Mean', 'Bwd IAT Mean',
    'SYN Flag Cnt', 'ACK Flag Cnt', 'FIN Flag Cnt',
    'Label'  # Keeping the label for classification
]

# Ensure that only the selected features exist in the dataset
available_features = [col for col in reduced_features if col in data.columns]

# Reduce dataset to only the selected features
reduced_data = data[available_features]

reduced_data.head()
```

```python
#Plot histograms for features
reduced_data.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#Box plots
reduced_data.plot(kind='box', subplots=True, layout=(5,5), figsize=(15,10))
plt.show()
```

```python
#Add feature BytesPerPacket: Helps detect irregular packet sizes.
reduced_data['BytesPerPacket'] = data['Flow Byts/s'] / reduced_data['Flow Pkts/s']
reduced_data.head()
```

```python
#Add feature FlowRate & PacketRate: Identifies traffic anomalies, useful for DDoS detection.
reduced_data['FlowRate'] = data['Flow Byts/s'] / data['Flow Duration']
reduced_data['PacketRate'] = data['Flow Pkts/s'] / data['Flow Duration']
reduced_data.head()
```

```python
#Add features ByteRatio & PacketToByteRatio: Measures the ratio of sent/received data to detect unusual communication.
reduced_data['ByteRatio'] = data['Flow Byts/s'] / data['TotLen Fwd Pkts']
reduced_data['PacketToByteRatio'] = data['Flow Pkts/s'] / data['TotLen Fwd Pkts']
```

```python
#Apply Log Transformations: Applies to Duration and Packets to reduce skewness.
reduced_data['Flow Duration'] = np.log1p(reduced_data['Flow Duration'])
reduced_data['Tot Fwd Pkts'] = np.log1p(reduced_data['Tot Fwd Pkts'])
```

```python
#Add feature Packet Size Variance: Detects abnormal variations in traffic flow.
reduced_data["Packet_Size_Variance"] = data["TotLen Fwd Pkts"] - data["TotLen Bwd Pkts"]
```

## Print the features after adding all the features and applying transformations

```python
#Split the dataset to training and testing datasets
from sklearn.model_selection import train_test_split
train = reduced_data.sample(frac=0.8, random_state=42)
test = reduced_data.drop(train.index)
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Remove infinity and NaN values
train = train.replace([np.inf, -np.inf], np.nan).dropna()
test = test.replace([np.inf, -np.inf], np.nan).dropna()

# Split features and target
X_train = train.drop(columns=['Label'])
y_train = train['Label']
X_test = test.drop(columns=['Label'])
y_test = test['Label']

# Standardize the data (except for Decision Tree)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Logistic Regression Model
# Cell 3: Train Logistic Regression model
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)

# Predict and evaluate
y_pred_log_reg = log_reg.predict(X_test_scaled)
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
print(f"Logistic Regression Accuracy: {accuracy_log_reg:.4f}")
```

```python
#SVM Model
svm = SVC()
svm.fit(X_train_scaled, y_train)

# Predict and evaluate
y_pred_svm = svm.predict(X_test_scaled)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Accuracy: {accuracy_svm:.4f}")
```

```python
#Decision Tree Model
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)  # Decision trees don't need scaling

# Predict and evaluate
y_pred_decision_tree = decision_tree.predict(X_test)
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
print(f"Decision Tree Accuracy: {accuracy_decision_tree:.4f}")
```

```python
#Import classification report and confusion matrix
from sklearn.metrics import classification_report, confusion_matrix

def calculate_fpr(y_true, y_pred):
    """Calculate False Positive Rate (FPR) from the confusion matrix."""
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    fpr = fp / (fp + tn)  # False Positive Rate
    return fpr
```

```python
# Classification reports for each model

# Classification Report for Logistic Regression
print("Logistic Regression Report:")
print(classification_report(y_test, y_pred_log_reg))
fpr_log_reg = calculate_fpr(y_test, y_pred_log_reg)
print(f"False Positive Rate (FPR): {fpr_log_reg:.4f}\n")

# Classification Report for SVM
print("SVM Report:")
print(classification_report(y_test, y_pred_svm))
fpr_svm = calculate_fpr(y_test, y_pred_svm)
print(f"False Positive Rate (FPR): {fpr_svm:.4f}\n")

# Classification Report for Decision Tree
print("Decision Tree Report:")
print(classification_report(y_test, y_pred_decision_tree))
fpr_decision_tree = calculate_fpr(y_test, y_pred_decision_tree)
print(f"False Positive Rate (FPR): {fpr_decision_tree:.4f}\n")
```

```python
#Most important features

# Cell 10: Feature Importance Analysis for Decision Tree
import matplotlib.pyplot as plt

# Get feature importance from Decision Tree model
feature_importance = decision_tree.feature_importances_

# Create a DataFrame for better visualization
feature_importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importance
}).sort_values(by='Importance', ascending=False)

# Display the feature importance ranking
print("Feature Importance Ranking for Botnet Detection:")
print(feature_importance_df)

# Plot feature importance
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Feature Importance')
plt.ylabel('Feature Name')
plt.title('Decision Tree Feature Importance for Botnet Detection')
plt.gca().invert_yaxis()  # Highest importance at the top
plt.show()
```
